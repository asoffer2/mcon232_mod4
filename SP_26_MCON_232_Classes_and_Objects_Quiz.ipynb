## Short reminders before you begin
A **class** is a blueprint.  
An **object** is one instance created from that blueprint.  
The `__init__` method builds the starting state of each object.  
An **instance attribute** belongs to one object, usually written with `self.attribute_name`.  
A **class attribute** belongs to the class itself and is shared unless shadowed by an instance.  
A **method** is a function inside a class.  
When an object is passed into a method, the method receives a reference to that object.  
If the method **mutates the object's attributes**, the caller sees those changes after the method ends.  
If the method **reassigns the parameter variable** to a different object, that reassignment only changes the local variable inside the method.

### Question C1 — Create a simple class

Create a class named `Book`.

#### Requirements
- The constructor should accept:
  - `title`
  - `author`
  - `pages`
- Store each value in an instance attribute.
- Then create an object named `b1` with your own sample data.
- Print the title and author on one line.
- use `__str__`

**Reminder:** Each object should store its own instance data.

```python

# Write your answer here
class Book:
  def __init__(self, title: str, author: str, pages: int)-> None:
    '''initializes the object and stores attributes'''
    self.title = title
    self.author = author
    self.pages = pages

  def __str__(self) -> str:
    '''returns title and author on one line'''
    return f"Title: {self.title}, Author: {self.author}"

b1 = Book("Little Women", "Louisa May Alcott", 400)
print(b1)

```

### Question C4 — One object uses another object

Create two classes: `Player` and `Coach`.

#### `Player`
- constructor should accept:

#### `Coach`
- add a method `praise_player(player_obj)` that returns a string like:

Create a player object and a coach object, then call the method.

**Reminder:** This question practices passing an object into another object's method and using its data.

```python

# Write your answer here
class Player:
  def __init__(self, name:str, points:int) -> None:
    '''initializes the object and stores attributes'''
    self.name = name
    self.points = points

class Coach:
  def praise_player(self, player: Player) -> str:
    '''returns the praise string using specific player data'''
    return f"Great job, {player.name}! You scored {player.points} points."


player = Player("Ella", 40)
coach = Coach()
print(coach.praise_player(player))
```

### Question C5 — Mutation vs reassignment

Create two classes: `Pet` and `Groomer`.

#### `Pet`
- constructor should accept:
  - `name`
  - `clean`

#### `Groomer`
Add two methods:

1. `wash_pet(pet_obj)`  
   - set `pet_obj.clean` to `True`

2. `replace_pet(pet_obj)`  
   - reassign `pet_obj` to a brand new `Pet("Temporary", True)`

#### After writing the classes
- Create one pet object that starts dirty.
- Call `wash_pet(...)`
- Print the original pet's `name` and `clean` value
- Then call `replace_pet(...)`
- Print the original pet's `name` and `clean` value again

#### Finally, answer in a comment:
- Which method changed the original object?
- Which method did not?
- Why?

**Reminder:** Mutation changes the object itself. Reassignment only changes the local parameter name.

```python

# Write your answer here
class Pet:
  def __init__(self, name: str, clean: bool) -> None:
    '''initializes the object and stores attributes'''
    self.name = name
    self.clean = clean

  def __str__(self) -> str:
    return f"Name: {self.name}, Clean: {self.clean}"

class Groomer:
  def wash_pet(self, pet_obj: Pet) -> None:
    pet_obj.clean = True

  def replace_pet(self, pet_obj: Pet) -> None:
    pet_obj = Pet("Temporary", True)

my_pet = Pet("Cocoa", False)
groomer = Groomer()
groomer.wash_pet(my_pet)
print(my_pet)
groomer.replace_pet(my_pet)
print(my_pet)

''' wash_pet changed the original object, while replace_pet did not. Wash_pet mutated the value of an object which is allowed sinced it is passed by object reference, while replace_pet reassigned the variable which will work only locally.'''
```
