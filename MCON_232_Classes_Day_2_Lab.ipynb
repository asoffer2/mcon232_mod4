```python
class Pet:
  country = "USA" # will be a class attribute
  def __init__(self, name: str, age:int, owner:str) -> None:
    self.name = name
    self.age = age
    self.owner = owner

# getters
  def get_name(self):
    return self.name

  def get_age(self):
    return self.age

  def get_owner(self):
    return self.owner

# setters
  def set_name(self, name):
    self.name = name

  def set_age(self, age):
    self.age = age

  def set_owner(self, owner):
    self.owner = owner

pet1 = Pet("muppet", 5, "Me")
pet2 = Pet("Polly", 3, "Me")
print(pet1.get_name()) # now its a method, dont forget parenthesis
print(pet2.name)
print(pet2.set_name("Ella"))
Pet.country = "EU"
print(pet1.country)
# if ur object shadows it, value will reflect the more specific one
pet1.country = "Italy" # more specific so will hold even if generally changed
Pet.country = "Israel"
```
